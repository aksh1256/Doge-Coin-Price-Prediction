```python
# Description: This program predicts the price of  dogecoin for next 30 days
```

```python
# Import the libraries
import pandas as pd
import numpy as np
```

```python
# Load the data
df = pd.read_csv("doge.csv")
```

```python
# Show the first 7 rows of data
df.head(7)
```

```python
#remove the date column
df.drop(['date'],1,inplace=True)
```

```python
# Show the first 7 rows of new dataset
df.head(7)
```

```python
# A variable for predicting 'n' daysout into the future
prediction_days = 30 #n = prediction_days = 30

# Create another column shifted 'n' units up
df['Prediction'] = df[['value']].shift(-prediction_days)
```

```python
#Show the first 7 rows of the new dataset
df.head(7)
```

```python
# Show the last 7 rows of new dataset
df.tail(7)
```

```python
# Create the independent dataset
# convert the dataframe to a numpy array and drop the prediction column
X = np.array(df.drop(['Prediction'], 1))

# Remove the last 'n' rows where 'n' is the prediction_days
X = X[:len(df)-prediction_days]

print(X)
```

```python
# Create the dependent dataset

# Convert the dataframe to a numpy array
y = np.array(df['Prediction'])

# Get all the values except the last 'n' rows 
y = y [:-prediction_days]

print(y)
```

```python
#Split the data into 80% of training and 20% of testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
#Set the prediction_days_array equal to the last 30 rows from the original dataset
prediction_days_array = np.array(df.drop(['Prediction'],1))[-prediction_days:]

print(prediction_days_array)
```

```python
from sklearn.svm import SVR

#Create and train the Support Vector Machine (Regression) using radial basis function
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)
svr_rbf.fit(x_train, y_train)
```

```python
#Test the model
svr_rbf_confidence = svr_rbf.score(x_test, y_test)

print("svr_rbf accuracy: ", svr_rbf_confidence)
```

```python
#Print the predicted values
svm_prediction = svr_rbf.predict(x_test)
print(svm_prediction)

print()

# Print the actual values
print(y_test)
```

```python
# Print the model prediction for the next 'n=30' days
svm_prediction = svr_rbf.predict(prediction_days_array)
print(svm_prediction)

print()

# Print the actual values for the dogecoin for the last 30 days
print(df.tail(prediction_days))
```
